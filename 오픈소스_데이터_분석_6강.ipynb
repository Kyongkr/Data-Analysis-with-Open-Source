1-2 Python을 사용하여 API를 호출하는 코드를 작성하고, 2015년 1월부터 2024년 12월까지의 개
인 유형의 현년 전기, 가스, 수도, 지역난방 에너지 사용량 데이터를 수집하는 프로그램을 작성하
시오. API 호출 성공을 확인할 수 있는 실행 결과를 캡처하여 첨부하시오. (7점)

```python
import requests
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')


# 1-2
# 기초값 설정
url = "http://openapi.seoul.go.kr:8088/{key}/json/energyUseDataSummaryInfo/1/10/{year}/{month}"
api_key = "58444d444a6f75743731714964444e"
year_start, year_end = 2015, 2024

# 데이터 수집
data_list = []

print(f"서울시 개인 에너지 사용량 수집({year_start}.01 ~ {year_end}.12)")
print("-" * 41)


try:

    # 2015년 ~ 2024년까지
    for year in range(year_start, year_end + 1):
        for month in range(1, 13):

            # api_url 포맷에 사용할 월 문자
            month_str = str(month).zfill(2)

            # API 호출 URL
            api_url = url.format(
                key=api_key,
                year=year,
                month=month_str
            )

            print(f"       -> 호출 중: {year}년 {month:02d}월 ~ ", end="")

            # API 호출
            response = requests.get(api_url)

            # 응답 상태 코드 확인
            if response.status_code == 200:
                # JSON 데이터 로드
                data = response.json()

                # API 호출 성공 여부 확인
                if 'energyUseDataSummaryInfo' in data:
                    rows = data['energyUseDataSummaryInfo']['row']
                    personal_data = next((row for row in rows if row.get('MM_TYPE') == '개인'), None)

                    if personal_data:
                        data_list.append({
                            '날짜': year * 100 + month,
                            '전기': personal_data.get('EUS'),
                            '가스': personal_data.get('GUS'),
                            '수도': personal_data.get('WUS'),
                            '지역난방': personal_data.get('HUS'),
                            })
                        print("성공")

                    else:
                        print("개인 유형 없음")

            else:
                print("API 호출 실패")

            time.sleep(0.1)

except requests.exceptions.RequestException:
    print("API 요청 실패")
```

하는 코드와 실행 결과를 첨부하시오.

```python
df = pd.DataFrame(data_list)
df.info()
```

2-2 연도별, 계절별 분석을 위해 날짜 컬럼을 활용하여 연도(year)와 계절(season) 컬럼을 추가하
는 전처리 코드를 작성하고, 변환 결과를 확인할 수 있는 출력 결과를 첨부하시오.
※ 계절 구분: 봄(3-5월), 여름(6-8월), 가을(9-11월), 겨울(12-2월) (4점)

```python
# '연도' 칼럼
df['연도'] = df['날짜'].astype(str).str[:4]

# 계절의 조건과 결과를 정의
month_num = df['날짜'] % 100

conditions = [
    (month_num >= 3) & (month_num <= 5), # 봄
    (month_num >= 6) & (month_num <= 8), # 여름
    (month_num >= 9) & (month_num <= 11) # 가을
]
choices = ['봄', '여름', '가을']

# '계절' 칼럼 추가, 나머지는 '겨울'
df['계절'] = np.select(conditions, choices, default='겨울')
df.head()
```

3-1 연도별 에너지 사용 총 사용량(전기+가스+수도+지역난방) 변화량을 선 그래프로 시각화하고,
그래프에 자신의 학번 뒤 4자리를 제목에 포함하여 저장하시오. (예: "연도별 에너지 사용 총액 변
화 - 1234") 시각화 코드와 생성된 그래프를 첨부하시오. (4점)

```python
# 에너지 사용량 칼럼을 수치형 변환
numeric_cols = ['전기', '가스', '수도', '지역난방']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# 연도별 총 사용량
df['총 사용량'] = df[numeric_cols].sum(axis=1)

# '연도'별로 '총 사용량'을 합산, 새 데이터프레임을 생성
annual_df = df.groupby('연도')['총 사용량'].sum().reset_index()
annual_df.columns = ['연도', '연간 총 사용량']

# X축: 연도, Y축: 연간 총 사용량
plt.figure(figsize=(10, 6))
plt.plot(annual_df['연도'], annual_df['연간 총 사용량'], marker='o', linestyle='-', color='b')

# 제목 설정
plt.title('2015년~2024년 연도별 개인 에너지 총 사용량 변화 - 6989', fontsize=15)
plt.xlabel('연도', fontsize=12)
plt.ylabel('연간 총 에너지 사용량', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# X축, Y축 표
plt.xticks(annual_df['연도'], rotation=0)
plt.ticklabel_format(axis='y', style='plain')
plt.show()
```

3-2 계절별 가스 사용량 평균을 막대 그래프로 시각화하고, 각 막대에 구체적인 수치를 표시하시
오. 시각화 코드와 생성된 그래프를 첨부하시오. (4점)

```python
# 계절별 가스 사용량 평균 계산
season_gas_usage_average = df.groupby('계절')['가스'].mean().reindex(['봄', '여름', '가을', '겨울'])

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(season_gas_usage_average.index, season_gas_usage_average.values, color=['yellow', 'red', 'blue', 'grey])

# 그래프 제목 및 라벨 설정
plt.title('계절별 평균 가스 사용량', fontsize=15)
plt.xlabel('계절', fontsize=12)
plt.ylabel('평균 가스 사용량', fontsize=12)
plt.ticklabel_format(axis='y', style='plain')

# 각 막대 위에 구체적인 수치 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}',
             ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
